```python
import random
PUPILS_K5 = 4197.19
PUPILS_68 = 4533.31
PUPILS_912 = 4322.70
POVERTY = 503.66
INTERVENTION = 10071.6
LANGUAGE_LEARNER = 9486.27
SPECIAL_ED = 31579.26
CTE = 2517.67
SPECIALIZED_ACADEMIC = 2517.68
pupils = [0, 0, 1000]
poverty = False

actual_budget = 4003095.17

def school(pupils, poor, percent_intervention, percent_ell, percent_special_ed, percent_cte, percent_specialized):
    poverty_funding = poverty*sum(pupils)*POVERTY
    base_funding = PUPILS_K5*pupils[0] + PUPILS_68*pupils[1] + PUPILS_912*pupils[2]
    intervention_funding = INTERVENTION*sum(pupils)*percent_intervention
    language_learner_funding = LANGUAGE_LEARNER*sum(pupils)*percent_ell
    special_education_funding = SPECIAL_ED*sum(pupils)*percent_special_ed
    cte_funding = CTE*sum(pupils)*percent_cte
    specialized_funding = SPECIALIZED_ACADEMIC*sum(pupils)*percent_specialized
    total_funding = (base_funding + poverty_funding + intervention_funding + language_learner_funding + special_education_funding + cte_funding + specialized_funding)
    return(total_funding)

for i in range(0, 10000):
    percentage1 = random.randint(0, 100)/100.0 #academic intervention    
    percentage2 = random.randint(0, 100)/100.0 #english language learner
    percentage3 = random.randint(0, 100)/100.0 #special education
    percentage4 = random.randint(0, 100)/100.0 #CTE
    percentage5 = random.randint(0, 100)/100.0 #specialized academics
    budget = school(pupils, poverty, percentage1, percentage2, percentage3, percentage4, percentage5)
    if budget >= actual_budget and budget <= 20000000:
        print(percentage1, percentage2, percentage3, percentage4, percentage5)
        print(budget)
```
